```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    id:int
    name:str
    age:int
    email:str
    
 
```

```python
user_data = {
    "id": "101",
    "name": "Asif",
    "age": "21",
    "email": "asif@gmail.com"
}   
user1=User(**user_data)
print(user1)
```

```python
from pydantic import BaseModel, Field
```

```python
class Person(BaseModel):
    name:str=Field(min_length=3)
    age:int=Field(ge=18)
    marks:float=Field(gt=0, le=100)
    email:str=Field(min_length=5, pattern="^[a-zA-Z0-9]+@")
```

```python
p1=Person(
    name="asif ali",
    age=20,
    marks=100,
    email="user@gamil.com"
)
print(p1)
```

```python
class Address(BaseModel):
    city:str
    country:str

class Customer(BaseModel):
    name:str
    address:Address    
```

```python
user_data={
    "name": "asif ali",
    "address":{
        "city":"Islamabad",
        "country":"pakistan"
    }
}
c1=Customer(**user_data)
```

```python
print(c1)
```

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True
```

```python
emp=Employee(id=11, name="asifali")
print(emp)
```
