```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
print(x_train.min(), x_train.max())
print(x_test.min(), x_test.max())
print(x_train.dtype)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```
